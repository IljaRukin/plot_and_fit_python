```python
%matplotlib inline
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
x=y=np.linspace(-10*np.pi,10*np.pi,100)
X,Y = np.meshgrid(x,y)
Z= np.multiply(np.exp(-np.abs(X)/20),np.sin(Y/10))
```

```python
fig = plt.figure(figsize=(5,4), dpi=80)
ax = fig.gca(projection='3d')

#contour
forcolorbar1 = ax.contour(X, Y, Z, np.linspace(-1,1,21), cmap=cm.coolwarm, alpha=1)
fig.colorbar(forcolorbar1, shrink=0.8, aspect=10)

plt.show()
```

```python
fig = plt.figure(figsize=(5,4), dpi=80)
ax = fig.gca(projection='3d')

#contour thick lines
forcolorbar1 = ax.contour(X, Y, Z, np.linspace(-1,1,21), cmap=cm.coolwarm, extend3d=True, alpha=0.8)
fig.colorbar(forcolorbar1, shrink=0.8, aspect=10)

plt.show()
```

```python
fig = plt.figure(figsize=(5,4), dpi=80)
ax = fig.gca(projection='3d')

#contour
forcolorbar1 = ax.contourf(X, Y, Z, np.linspace(-1,1,21), cmap=cm.coolwarm, alpha=0.8)
fig.colorbar(forcolorbar1, shrink=0.8, aspect=10)

plt.show()
```

```python
fig = plt.figure(figsize=(5,4), dpi=80)
ax = fig.gca(projection='3d')

#contour projections to some axis
ax.contour(X, Y, Z, np.linspace(-1,1,11), zdir='z', offset=-1, cmap=cm.coolwarm)
ax.contour(X, Y, Z, np.linspace(-10*np.pi,10*np.pi,11), zdir='x', offset=-10*np.pi, cmap=cm.coolwarm)
ax.contour(X, Y, Z, np.linspace(-10*np.pi,10*np.pi,21), zdir='y', offset=10*np.pi, cmap=cm.coolwarm)

plt.show()
```

```python
fig = plt.figure(figsize=(5,4), dpi=80)
ax = fig.gca(projection='3d')

#thick contour projections to some axis
ax.contourf(X, Y, Z, np.linspace(-1,1,11), zdir='z', offset=-1, cmap=cm.coolwarm)
ax.contourf(X, Y, Z, np.linspace(-10*np.pi,10*np.pi,11), zdir='x', offset=-10*np.pi, cmap=cm.coolwarm)
ax.contourf(X, Y, Z, np.linspace(-10*np.pi,10*np.pi,21), zdir='y', offset=10*np.pi, cmap=cm.coolwarm)

plt.show()
```
