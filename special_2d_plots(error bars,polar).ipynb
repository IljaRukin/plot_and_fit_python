```python
%reset -f
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2*np.pi,20)
y = np.sin(x)
xerror = 0.2
yerror = 0.1*np.cos(x)
```

```python
fig = plt.figure(figsize=(5,4), dpi=80)
ax = plt.subplot(1,1,1)

#contour
ax.errorbar(x, y, xerr=xerror, yerr=yerror, fmt='o', markersize=2, capsize=5)


plt.show()
```

```python
phi=np.linspace(0,2*np.pi,20)
r = np.cos(phi)**2
xerror = 0.2
yerror = 0.1*np.cos(phi)
```

```python
fig = plt.figure(figsize=(5,4), dpi=80)
ax = fig.gca(projection='polar')

#contour
ax.plot(phi, r)
ax.errorbar(phi, r, xerr=xerror, yerr=yerror, fmt='o', markersize=2, capsize=0)

plt.show()
```
