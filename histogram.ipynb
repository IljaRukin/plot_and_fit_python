```python
#multiple bar plots
%reset -f
%matplotlib notebook
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

#data
x = np.linspace(0,10,11)
data = np.random.rand(4,11)

fig = plt.figure(figsize=(8,6), dpi=80)
ax = fig.add_subplot(111, projection='3d')

colors = ['r', 'g', 'b', 'y']
y = [3,2,1,0]

for i in range(0,4):
    ax.bar(x, data[i], y[i], zdir='y', color=colors[i], alpha=0.8)

plt.show()
```

```python
#3d bar plot
%reset -f
%matplotlib notebook
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

#data
xlen=10;ylen=10
data = np.random.rand(xlen,ylen)
xx,yy = np.meshgrid(np.arange(0,xlen),np.arange(0,ylen))
x=xx.flatten()
y=yy.flatten()
z=data.flatten()

fig = plt.figure(figsize=(8,6), dpi=80)
ax = fig.add_subplot(111, projection='3d')

#custom color
forcolor = cm.jet(z)

ax.bar3d(x,y,0, 0.6,0.6,z, color=forcolor)

plt.show()
```

```python
#histogram (unordered data accummulated in steps)
%reset -f
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#unordered data points
#gauss
mu, sigma = 0, 1 #mean and standard deviation
data = np.random.normal(mu, sigma, 1000)

fig = plt.figure(figsize=(5,4), dpi=80)
ax = plt.subplot()

ax.hist(data, range=(-5,5), bins=20, color='blue', edgecolor='k')
#optional for probability: density=True

plt.show()
```

```python
#histogram (unordered data cummulated in steps)
%reset -f
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

#unordered data points
#random
data0 = np.random.randint(-5,5,1000)
#gauss
mu, sigma = 0, 3 #mean and standard deviation
data1 = np.random.normal(mu, sigma, 1000)

fig = plt.figure(figsize=(5,4), dpi=80)
ax = plt.subplot()

ax.hist((data0,data1), range=(-5,5), bins=10, color=('blue','red'), edgecolor='k')

plt.show()
```

```python
#2d histogram (unordered data accummulated in steps)
%reset -f
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#unordered data points
#gauss
mu, sigma = 0, 1 #mean and standard deviation
data1 = np.random.normal(mu, sigma, 1000)
data2 = np.random.normal(mu, sigma, 1000)

fig = plt.figure(figsize=(5,4), dpi=80)
ax = plt.subplot()

ax.hist2d(data1,data2, range=((-5,5),(-5,5)), bins=20, cmap='jet', edgecolor='k')

plt.show()
```
