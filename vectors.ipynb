```python
%reset -f
#%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

#x,y,z grid
x, y = np.meshgrid(np.linspace(-1,1,20),
                     np.linspace(-1,1,20))

#arrow direction
u = x
v = y

#for color mapping
r2=np.multiply(u,u)+np.multiply(v,v)

#plot
fig = plt.figure(figsize=(8,6), dpi=80)
ax = plt.subplot()

ax.quiver(x, y, u, v, r2, cmap='jet', scale=20, linewidth=0.5)#, edgecolor='k', facecolor='None')

plt.show()
```

```python
%reset -f
#%matplotlib notebook
%matplotlib inline
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

#x,y,z grid
x, y, z = np.meshgrid(np.linspace(-1,1,10),
                     np.linspace(-1,1,10),
                     np.linspace(-1,1,3))
#x,y,z-shape: (10, 10, 3)

#arrow direction
u = x
v = y
w = z

#for color mapping
r2=np.multiply(u,u)+np.multiply(v,v)+np.multiply(w,w)
forcolor = r2.flatten()#1d array

#plot
fig = plt.figure(figsize=(10,8), dpi=80)
ax = plt.subplot(projection='3d')

#ax.quiver(x, y, z, u, v, w, cmap='Reds', length=0.2)
#optional (same length): normalize=True

q = ax.quiver(x, y, z, u, v, w, cmap='jet', length=0.1, lw=2)
q.set_array(forcolor)

plt.show()
```
