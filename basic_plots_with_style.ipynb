```python
#2d plotting with style and subplots
%reset -f
#%matplotlib notebook
%matplotlib inline

#import library
#plotting
import matplotlib.pyplot as plt
#color
from matplotlib import cm
from matplotlib import colors

#compute data points for plot
import numpy as np
x = np.linspace(-2*np.pi,2*np.pi,100)
y = np.linspace(-2*np.pi,2*np.pi,100)
xx,yy=np.meshgrid(x,y)
zz = np.sin(xx) + np.cos(yy)

xx = np.reshape(xx,(np.product(np.shape(xx)),1))
yy = np.reshape(yy,(np.product(np.shape(yy)),1))
zz = np.reshape(zz,(np.product(np.shape(zz)),1))


fig = plt.figure(num=None, figsize=(10,5), dpi=80, facecolor='w', edgecolor='k')
ax0=plt.subplot(1,1,1) #2x1 subplots, select 1 plot


#custom color
colormap = plt.cm.jet
normalize = colors.Normalize(vmin=-2, vmax=2)

#plot
sc = plt.scatter(xx, yy, c=zz, s=20, cmap=colormap, norm=normalize)
plt.colorbar(sc)

#labels
ax0.set_title('Function', fontsize=16)
ax0.set_xlabel('X', fontsize=12)
ax0.set_ylabel('Y', fontsize=12)
ax0.tick_params(axis='both', which='major', labelsize=16)
#title space
tt0 = ax0.title
tt0.set_position([.5, 1.05])

#dimensions & ticks
ax0.set_xticks(np.linspace(-2*np.pi,2*np.pi,5))
ax0.set_xticklabels((r'$-2\pi$',r'$-\pi$',0,r'$\pi$',r'$2\pi$'))
ax0.set_yticks(np.linspace(-2*np.pi,2*np.pi,5))
ax0.set_yticklabels((r'$-2\pi$',r'$-\pi$',0,r'$\pi$',r'$2\pi$'))
ax0.set_xlim(-2*np.pi,2*np.pi)
ax0.set_ylim(-2*np.pi,2*np.pi)
ax0.grid(False)
#ax0.axis('equal')

fig.gca().set_autoscale_on(False)
#rescale subplots
plt.tight_layout()

#show plot
plt.show()
#plt.draw()

#save image
#plt.savefig('plot.png')
```

```python
#3d line/scatter plotting with style
%reset -f
#%matplotlib notebook
%matplotlib inline

#import library
#plotting
import matplotlib.pyplot as plt
#3d plotting
from mpl_toolkits.mplot3d import Axes3D
#color
from matplotlib import cm

#compute data points for plot
import numpy as np
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

#setup figure
fig = plt.figure(num=None, figsize=(10,5), dpi=80, facecolor='w', edgecolor='k')
ax0=plt.subplot(1,1,1, projection='3d')


#custom color (for ax0.scatter)
c1=cm.jet(z)

#plot
ax0.plot(x,y,z-0.5,color="blue", linewidth=4, label='function')
ax0.scatter(x,y,z+0.5, s=20, c=c1, marker='^', depthshade=True, label='function')

#labels
ax0.legend(fontsize=12, bbox_to_anchor=(0.8,0.9))
ax0.set_title('Function', fontsize=16)
ax0.set_xlabel('X', fontsize=12)
ax0.set_ylabel('Y', fontsize=12)
ax0.set_zlabel('Z', fontsize=12)
ax0.tick_params(axis='both', which='major', labelsize=16)

#dimensions & ticks
ax0.set_xticks([-4,-2,0,2,4])
ax0.set_yticks([-4,-2,0,2,4])
ax0.set_zticks([-4,-2,0,2,4])
ax0.set_xlim(-4, 4)
ax0.set_ylim(-4, 4)
ax0.set_zlim(-4, 4)
ax0.grid(True)
#ax0.axis('equal')
ax0.view_init(elev=20, azim=-35)

fig.gca().set_autoscale_on(False)
#rescale subplots
plt.tight_layout()

#show plot
plt.show()
#plt.draw()

#save image
#plt.savefig('plot.png')
```
