```python
#display image
%reset -f
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
import cv2

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.333, 0.333, 0.333])


#only png
#pic = plt.imread('example0.png') #350x525x4 array
#bw_pic = rgb2gray(pic) #350x525x1 array


#any image
BGR_pic = cv2.imread('example0.png')
pic = cv2.cvtColor(BGR_pic, cv2.COLOR_BGR2RGB)
bw_pic = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)


#plot
fig = plt.figure(figsize=(12,5), dpi=150)

plt.subplot(3,1,1)
plt.imshow(pic)
#rgba image (no cmap and colorbar)
plt.colorbar()
#plt.axis('off')

plt.subplot(3,1,2)
plt.imshow(bw_pic, cmap='gray')
plt.colorbar()
#plt.axis('off')

plt.subplot(3,1,3)
plt.imshow(bw_pic, cmap='jet')
plt.colorbar()
#plt.axis('off')

plt.show
```
